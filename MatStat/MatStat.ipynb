```python
import matstat

import numpy as np
```

```python
DVS = matstat.discreetVariationSeries
IVS = matstat.intervalVariationSeries
```

```python
rng = np.random.default_rng()
```

```python
disc1_dist = DVS(disc1)
```

```python
disc1_dist.plot_polygon()
```

# Interval series

## Building from raw collection of data

```python
cont = rng.logistic(10, 15, size=30)
```

```python
ivs = IVS(cont)
```

```python
ivs.table
```

### Frequency polygon

```python
ivs.plot_polygon()
```

### Empirical distribution function (Fx)

```python
ivs.plot_Fx()
```

### Frequency histograms

```python
ivs.plot_hist('x', 'n/h')
```

```python
ivs.plot_hist('x', 'w/h')
```

## Building from ranges and distribution

```python
lims = sorted(rng.choice(np.arange(30, 100), replace=False, size=(6)))
ns = rng.choice(np.arange(0, 200), replace=False, size=(5))
```

```python
cont_data1 = np.array([(left, right, n) for left, right, n in zip(lims[:-1], lims[1:], ns)])
```

```python
ivs1 = IVS(cont_data1)
```

```python
ivs1.table
```

```python
ivs1.show_stats()
```

### Frequency polygon

```python
ivs1.plot_polygon()
```

### Empirical distribution function (Fx)

```python
ivs1.plot_Fx()
```

### Frequency histograms

```python
ivs1.plot_hist('x', 'n/h')
```

```python
ivs1.plot_hist('x', 'w/h')
```
