```python
from sympy import *
import numpy as np

import matplotlib.pyplot as plt
```

```python
def plot_func_and_der(x='x', y=Symbol('x')**2, x_range=(-10, 10)):
    yprime = y.diff(x)
    
    f = lambdify(x, y, 'numpy')
    fder = lambdify(x, yprime, 'numpy')
    
    xs = np.linspace(x_range[0], x_range[1], 100)
    
    if type(fder(xs)) != np.ndarray:
        fders = [fder(xs) for i in range(len(xs))]
    else:
        fders = fder(xs)
    
#     return f(xs)
    plt.plot(xs, f(xs), label=f'f(x) = {y}')
    plt.plot(xs, fders, label=f"f\'(x) = {yprime}")

    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.ylim(-50, 50)
    
    plt.grid()
    plt.legend()
```

```python
plot_func_and_der(x='x', y=exp((Symbol('x'))/3))
```

```python
plot_func_and_der(x='x', y=2*(Symbol('x') + exp(Symbol('x'))))
```
